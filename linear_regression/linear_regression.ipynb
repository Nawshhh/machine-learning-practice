### Read Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url = 'https://raw.githubusercontent.com/Nawshhh/machine-learning-practice/refs/heads/main/linear_regression/spotify_churn_dataset.csv'
spotify_df = pd.read_csv(url)
```

```python
spotify_df
```

### Check for missing values

```python
print(spotify_df.isnull().sum())
```

```python
print(spotify_df['ads_listened_per_week'].unique())
```

```python
spotify_df = spotify_df[['listening_time','songs_played_per_day','ads_listened_per_week','skip_rate']].copy()
print(spotify_df)
```

### Plot the relationships

```python
spotify_df.plot(x='ads_listened_per_week',y='skip_rate', kind='bar', title='Relationship of ads and skip rate')
plt.ylabel('Values')
plt.show()
```
